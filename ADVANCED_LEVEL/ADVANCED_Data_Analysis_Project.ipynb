```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(
    r"C:\Users\hp\Favorites\Downloads\sales_data_sample.csv",
    encoding="latin1"
)

df.head()
```

```python
## Dataset Overview
df.info()
df.describe()
```

```python
## Data Cleaning
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
## Sales Trend Over Time
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
```

```python
#Line Plot (Sales Trend)
sales_trend = df.groupby(df['ORDERDATE'].dt.year)['SALES'].sum()

plt.plot(sales_trend.index, sales_trend.values)
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.title("Yearly Sales Trend")
plt.show()
```

```python
## Product Line Analysis
#Bar Chart (Product Line)
product_sales = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False)

product_sales.plot(kind='bar')
plt.title("Sales by Product Line")
plt.xlabel("Product Line")
plt.ylabel("Total Sales")
plt.show()
```

```python
## Country-wise Sales Analysis
country_sales = df.groupby('COUNTRY')['SALES'].sum().sort_values(ascending=False).head(10)

country_sales.plot(kind='bar')
plt.title("Top 10 Countries by Sales")
plt.xlabel("Country")
plt.ylabel("Total Sales")
plt.show()
```

```python
## Quantity Ordered vs Sales
#Scatter Plot
plt.scatter(df['QUANTITYORDERED'], df['SALES'])
plt.xlabel("Quantity Ordered")
plt.ylabel("Sales")
plt.title("Quantity Ordered vs Sales")
plt.show()
```

```python
## Correlation Analysis
#Heatmap
plt.figure(figsize=(8,5))
sns.heatmap(df[['SALES','QUANTITYORDERED','PRICEEACH']].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
