```python
## 1. Library Overview

### Matplotlib
# Matplotlib is a popular Python library used for creating static and customizable visualizations.import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Line Plot Example (Matplotlib)")
plt.show()
```

```python
### Bar Chart
# Bar charts are used to compare values across different categories.
plt.bar(["A", "B", "C"], [10, 15, 7])
plt.title("Bar Chart Example (Matplotlib)")
plt.show()
```

```python
### Histogram
# Histograms show the distribution of numerical data.
import numpy as np

data = np.random.randn(1000)
plt.hist(data, bins=20)
plt.title("Histogram Example (Matplotlib)")
plt.show()
```

```python
### Seaborn
# Seaborn is built on top of Matplotlib and is mainly used for statistical visualizations.
import seaborn as sns
import pandas as pd

df = pd.DataFrame({
    "x": [1, 2, 3, 4],
    "y": [10, 20, 25, 30]
})

sns.scatterplot(x="x", y="y", data=df)
```

```python
### Box Plot
# Box plots help identify outliers and data spread.
sns.boxplot(data=data)
```

```python
### Pie Chart
# Pie charts show percentage contribution of categories.
plt.pie([40, 30, 20, 10],
        labels=["A", "B", "C", "D"],
        autopct="%1.1f%%")
plt.title("Pie Chart Example")
plt.show()
```
